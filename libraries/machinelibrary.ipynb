```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
x=[5,15,25,35,45,55]
y=[10,20,30,40,50,60]   
slope,intercept,r,p,std_err=stats.linregress(x,y)

def myfunc(x):
    return slope * x + intercept
mymodel=list(map(myfunc,x))
plt.scatter(x,y)
plt.plot(x,y)
plt.plot(x,mymodel)
plt.show()
print(r)
print(p)
```
