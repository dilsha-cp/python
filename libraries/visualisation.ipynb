```python
import matplotlib.pyplot as plt
import numpy as np

#Line plot
x=[1,2,3,4]
y=[11,22,33,44]

plt.plot(x,y, color='red',linestyle='-',marker='o',label='Line Plot')
plt.legend()
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[5,7,6,1,3,8,2,6,4,98]
y=[3,10,78,34,22,34,67,89,65,23]
plt.scatter(x,y,color="blue",label="scatter plot",marker="*")
plt.legend()
plt.title("Scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories=['A','B','C','D']
values=[10,20,30,40]
plt.bar(categories,values,color="red",label="bar plot")
plt.legend()
plt.title("Bar plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories=['A','B','C','D']
values=[10,20,30,40]
plt.barh(categories,values,color="red",label="bar plot")
plt.legend()
plt.title("Bar plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
labels = ['Apple','Banana','Mango','Orange','Pineapple','Watermelon']
sizes =[215,130,245,200,250,230]
colors = ['gold','yellowgreen','lightcoral','lightskyblue','red','green']
explode = (0.1,0,0,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
labels=['Apple','Banana','Mango','Orange']
sizes=[215,234,200,543]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
data=[[2,25,6,23],[1,4,3,2,5],[1,9,45,21]]
plt.boxplot(data,vert=True,patch_artist=True,labels=['Dataset 1','Dataset 2','Dataset 3'])
plt.title('Box Plot')
plt.xlabel('Dataset')
plt.ylabel('value')

plt.show()
```

```python
data = [ [2,3,5,8,12],[1,4,7,10,14],[0,2,6,8,13]]
plt.violinplot(data,showmeans=True,showmedians=True,showextrema=True)
plt.title('Violin Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
data=[[1,2,3,4],[5,6,7,8]]
plt.imshow(data,cmap='hot',interpolation='nearest')
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.show()
```
