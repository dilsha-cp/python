Q2:In which region maxium number of Confirmed cases were recorded?

```python
data.groupby('Country/Region').Confirmed.sum().sort_values(ascending=False)
```

```python
# Group by 'Country/Region' and find the sum of 'Confirmed' cases
grouped_data = data.groupby('Country/Region')[['Confirmed']].sum().reset_index()
# Find the region with the maximum confirmed cases
max_confirmed = grouped_data[grouped_data['Confirmed'] == grouped_data['Confirmed'].max()]

# Print the region with the maximum confirmed cases
print(f"Region with maximum confirmed cases: {max_confirmed['Country/Region'].values[0]}")

# Plot the choropleth map using Plotly
fig = px.choropleth(
    grouped_data, 
    locations='Country/Region', 
    locationmode='country names',  # Match country names
    color='Confirmed',  # Color based on confirmed cases
    hover_name='Country/Region',  # Display country names on hover
    color_continuous_scale='Reds',  # Color scale for confirmed cases
    title='COVID-19 Confirmed Cases by Region'
)

# Show the plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is already loaded into a DataFrame 'data'
# Group by 'Country/Region', sum the 'Deaths', then sort and take the 20 countries with minimum deaths
death_sum_by_region = data.groupby('Country/Region').Deaths.sum().sort_values(ascending=True).head(20)

# Create a pie chart
plt.figure(figsize=(10, 10))  # Adjust figure size
plt.pie(death_sum_by_region, labels=death_sum_by_region.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)

# Add a title
plt.title('Top 20 Countries with Minimum Deaths', fontsize=16)

# Show the plot
plt.show()

# Display the country names and death counts in tabular form
print(death_sum_by_region)
```

Q5:How many Confirmed,Death and Recovered cases were reported from india till 29 april 2020?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'data'
# Filter for India
india_data = data[data['Country/Region'] == 'India']

# Summarize the relevant data
summary = {
    'Deaths': india_data['Deaths'].sum(),
    'Recovered': india_data['Recovered'].sum()
}

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(summary.keys(), summary.values(), color=['red', 'green'])

# Add titles and labels
plt.title('COVID-19 Deaths and Recoveries in India', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

Q5:country with maximum recovery and less deth rathe

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (assuming it's already in a DataFrame called 'data')
# Example: data = pd.read_csv('your_data.csv')

# Group by 'Country/Region' and sum the relevant columns
country_data = data.groupby('Country/Region').agg({
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Confirmed': 'sum'
}).reset_index()

# Calculate Death Rate and Recovery Rate
country_data['Death Rate (%)'] = (country_data['Deaths'] / country_data['Confirmed']) * 100
country_data['Recovery Rate (%)'] = (country_data['Recovered'] / country_data['Confirmed']) * 100

# Find the country with maximum recovery rate and minimum death rate
target_country = country_data.loc[
    (country_data['Death Rate (%)'] == country_data['Death Rate (%)'].min()) &
    (country_data['Recovery Rate (%)'] == country_data['Recovery Rate (%)'].max())
]

# Extract the name of the target country
country_name = target_country['Country/Region'].values[0]
deaths = target_country['Deaths'].values[0]
recoveries = target_country['Recovered'].values[0]

# Print the target country and its metrics
print(f"Country with maximum recovery and minimum death rate: {country_name}")
print(f"Deaths: {deaths}, Recoveries: {recoveries}")

# Visualize the Death and Recovery data of that country
plt.figure(figsize=(8, 5))
plt.bar(['Deaths', 'Recovered'], [deaths, recoveries], color=['red', 'green'])

# Add titles and labels
plt.title(f'Death and Recovery Data for {country_name}', fontsize=16)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

Q6:Which regions have been most affected by COVID-19?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (assuming it's already in a DataFrame called 'data')
# Example: data = pd.read_csv('your_data.csv')

# Group by 'Country/Region' and sum the relevant columns
region_data = data.groupby('Country/Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum'
}).reset_index()

# Sort by Confirmed cases to focus on the top regions (optional)
top_regions = region_data.sort_values(by='Confirmed', ascending=False).head(10)

# Create pie chart for confirmed cases
plt.figure(figsize=(10, 10))
plt.pie(top_regions['Confirmed'], labels=top_regions['Country/Region'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 10 Regions by Confirmed COVID-19 Cases', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Create pie chart for deaths
plt.figure(figsize=(10, 10))
plt.pie(top_regions['Deaths'], labels=top_regions['Country/Region'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Top 10 Regions by COVID-19 Deaths', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Mortality Rate Analysis

```python
data['Mortality_Rate'] = (data['Deaths'] / data['Confirmed']) * 100
highest_mortality = data.nlargest(10, 'Mortality_Rate')
lowest_mortality = data.nsmallest(10, 'Mortality_Rate')

print("Highest Mortality Rates:")
print(highest_mortality[['Country/Region', 'Mortality_Rate']])

print("\nLowest Mortality Rates:")
print(lowest_mortality[['Country/Region', 'Mortality_Rate']])
```

Recovery Rate Analysis


```python
data['Recovery_Rate'] = (data['Recovered'] / data['Confirmed']) * 100
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Recovery_Rate', data=data)
plt.title('Recovery Rates by Country')
plt.xticks(rotation=90)
plt.show()
```
